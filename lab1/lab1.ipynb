```python
import numpy as np
import matplotlib.pyplot as plt 


# Exercitiul 1. Definirea functiilor
def x(t):
    return np.cos(520 * np.pi * t + np.pi / 3 )

def y(t):
    return np.cos(280 * np.pi * t - np.pi / 3 )

def z(t):
    return np.cos(120 * np.pi * t + np.pi / 3 )
```

```python
#(start, stop, how many evenly spaced points we will take)
num = np.linspace(0, 0.03, 60)

# subpunctele: a), b) -> Construirea si afisarea semnalelor
fig, axs = plt.subplots(3)
fig.suptitle("Semnale")

axs[0].plot(num, x(num))
axs[0].set_xlabel("timp")
axs[0].set_ylabel("Amplitudine")

axs[1].plot(num, y(num))
axs[1].set_xlabel("timp")
axs[1].set_ylabel("Amplitudine")

axs[2].plot(num, z(num))
axs[2].set_xlabel("timp")
axs[2].set_ylabel("Amplitudine")
plt.savefig("semnale_plot.pdf")
plt.show()
```

```python
#200Hz
num2 = np.linspace(0, 1, 200)
num3 = np.linspace(0, 1, 10000)

fig, axs = plt.subplots(3)
fig.suptitle("Esantionarea semnalelor 200Hz")

#Vector functii
funcs = [x, y, z]

#Afisam cele 3 functii esantionate pe 200Hz
for i in range(3):
    axs[i].plot(num3, funcs[i](num3))
    axs[i].stem(num2, funcs[i](num2))
    axs[i].set_xlabel("timp")
    axs[i].set_ylabel("Amplitudine")
    #Setam intervalul de afisare 0:0.1
    axs[i].set_xlim(0,0.1)
plt.savefig("esantionare_semnale_plot.pdf")
plt.show()
```

```python
#x[200Hz] pe intervalul 0:0.05

plt.plot(num3, x(num3))
plt.plot(num2, x(num2))
plt.stem(num2, x(num2))
plt.title("x[n] pe intervalul 0:0.05")
plt.xlim(0,0.05)
plt.xlabel("timp")
plt.ylabel("Amplitudine")
plt.savefig("esantionare_semnal_x[200Hz]_plot.pdf")
plt.show()
```

```python
# Exercitiul 2

# * Start Plot
fig, axs = plt.subplots(4, figsize=(10, 10))
plt.subplots_adjust(hspace=1)
fig.suptitle("Exercitiul 2")

# Semnal sinusoidal


def sin(f0, t):
    return np.sin(2 * np.pi * f0 * t)


# a) Semnal sinusoidal, frecnventa 400Hz, 1600 esantioane
f400 = 400
nt1 = np.linspace(0, 1, 100000)
nt2 = np.linspace(0, 1, 1600)

axs[0].plot(nt1, sin(f400, nt1))
axs[0].stem(nt2, sin(f400, nt2))
axs[0].set_title('Semnal sinusoidal, frecnventa 400Hz, 1600 esantioane')
axs[0].set_xlabel("timp")
axs[0].set_ylabel("Amplitudine")
axs[0].set_xlim(0, 0.01)

# b) Semnal sinusoidal, frecnventa 800Hz, 3 secunde
f800 = 800
t3 = np.linspace(0, 3, 100000)

axs[1].plot(t3, sin(f800, t3))
axs[1].set_title('Semnal sinusoidal, frecnventa 800Hz, 3 secunde')
axs[1].set_xlabel("timp")
axs[1].set_ylabel("Amplitudine")
axs[1].set_xlim(0, 0.01)

# c) Semnal sawtooth, 240Hz


def sawtooth(f0, t):
    return (f0 * t - np.floor(f0 * t + 1/2))


f240 = 240

axs[2].plot(nt1, sawtooth(f240, nt1))
axs[2].set_title('Semnal sawtooth, 240Hz')
axs[2].set_xlabel("timp")
axs[2].set_ylabel("Amplitudine")
axs[2].set_xlim(0, 0.012)

# d) Semnal square, 300Hz


def square(f0, t):
    return np.sign(np.sin(2 * np.pi * f0 * t))


axs[3].plot(nt1, square(300, nt1))
axs[3].set_title('Semnal square, 300Hz')
axs[3].set_xlabel("timp")
axs[3].set_ylabel("Amplitudine")
axs[3].set_xlim(0, 0.02)

# * Save plot
plt.savefig("exercitiul2.pdf")
plt.show()
```

```python
size = 128
random_array = np.random.rand(size, size)
colormaps = plt.colormaps()

colormaps = ['viridis', 'plasma', 'inferno', 'magma', 'cividis', 'coolwarm', 'jet', 'gray', 'autumn', 'spring']

# Create a single plot with subplots for each colormap
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for i, ax in enumerate(axes.ravel()):
    cmap = colormaps[i]
    ax.imshow(random_array, cmap=cmap)
    ax.set_title(cmap)
    ax.axis('off')

plt.tight_layout()
plt.savefig("random.pdf")
plt.show()
```
