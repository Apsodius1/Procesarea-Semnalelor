```python
# Exercitiul 2


def my_sinus(t, f = 1):
    return np.sin(2 * np.pi * t * f)


t_initial = np.linspace(0, 1, 1000)
t_sample = np.linspace(0, 1, 4)

fig, axs = plt.subplots(3)

axs[0].plot(t_initial, my_sinus(t_initial))
axs[0].stem(t_sample, my_sinus(t_sample))
axs[0].set_xlabel("timp")
axs[0].set_ylabel("Amplitudine")


axs[1].plot(t_initial, my_sinus(t_initial, 4))
axs[1].stem(t_sample, my_sinus(t_sample, 4))
axs[1].set_xlabel("timp")
axs[1].set_ylabel("Amplitudine")



axs[2].plot(t_initial, my_sinus(t_initial, 16))
axs[2].stem(t_sample, my_sinus(t_sample, 16))
axs[2].set_xlabel("timp")
axs[2].set_ylabel("Amplitudine")

plt.show()
```

```python
# Exercitiul 3


def my_sinus(t, f=1):
    return np.sin(2 * np.pi * t * f)


t_initial = np.linspace(0, 1, 1000)
t_sample = np.linspace(0, 1, 34)

fig, axs = plt.subplots(3)
fig.subplots_adjust(hspace=1)

axs[0].plot(t_initial, my_sinus(t_initial))
axs[0].stem(t_sample, my_sinus(t_sample))
axs[0].set_title("f=1")
axs[0].set_xlabel("timp")
axs[0].set_ylabel("Amplitudine")


axs[1].plot(t_initial, my_sinus(t_initial, 4))
axs[1].stem(t_sample, my_sinus(t_sample, 4))
axs[1].set_title("f=4")
axs[1].set_xlabel("timp")
axs[1].set_ylabel("Amplitudine")


axs[2].plot(t_initial, my_sinus(t_initial, 16))
axs[2].stem(t_sample, my_sinus(t_sample, 16))
axs[2].set_xlabel("timp")
axs[2].set_ylabel("Amplitudine")

axs[2].set_title("f= 13, 16")

axs[2].plot(t_initial, my_sinus(t_initial, 13))
axs[2].stem(t_sample, my_sinus(t_sample, 13), "r")
axs[2].set_xlabel("timp")
axs[2].set_ylabel("Amplitudine")

plt.show()
```

```python
# Exercitiul 5
fs, audio_data = scipy.io.wavfile.read("sounds/a.wav")


percentile = 1  
overlap_percent = 50  

overlap_samples = int(overlap_percent * fs / 100)
group_size = int(percentile * len(audio_data) / 100)  

# ! ValueError: noverlap must be less than nperseg.

frequencies, times, Sxx = scipy.signal.spectrogram(
    audio_data,
    fs,
    nperseg=group_size,
    noverlap=overlap_samples,
    window='hann'
)


plt.figure(figsize=(10, 6))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='auto')
plt.colorbar(label='Putere (dB)')
plt.title('Spectrograma Semnalului Audio')
plt.xlabel('Timp (secunde)')
plt.ylabel('Frecvență (Hz)')
plt.show()
```
