```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
import scipy
from timeit import default_timer as timer
```

```python
# Exercitiul 1 
N = [128, 256, 512, 1024, 2048, 4096, 8192]

t = np.linspace(0, 1, 10000)
semnal_sinus = np.sin(2 * np.pi * 110 * t)
my_times = []
numpy_times = []

for n_point in N:
    start = timer()
    X = np.zeros(n_point, dtype=np.complex_)
    for omega in range(n_point):
        for n in range(n_point):
            X[omega] += semnal_sinus[n] * np.exp(
                -2 * np.pi * 1j * omega * n / n_point
            )
    end = timer()
    my_times.append(end - start)

    start = timer()
    X = np.fft.fft(semnal_sinus, n_point)
    end = timer()
    numpy_times.append(end - start)
```

```python
fig, ax = plt.subplots()
ax.plot(N, my_times, label="DFT de mana")
ax.plot(N, numpy_times, label="numpy.fft.fft")
ax.set_xlabel("DFT points")
ax.set_ylabel("Timp (s)")
ax.legend()
ax.grid()
ax.set_yscale("log")

plt.plot()
plt.show()
```

```python
# Exercitiul 2


def my_sinus(t, f = 1):
    return np.sin(2 * np.pi * t * f)


t_initial = np.linspace(0, 1, 1000)
t_sample = np.linspace(0, 1, 4)

fig, axs = plt.subplots(3)

axs[0].plot(t_initial, my_sinus(t_initial))
axs[0].stem(t_sample, my_sinus(t_sample))
axs[0].set_xlabel("timp")
axs[0].set_ylabel("Amplitudine")


axs[1].plot(t_initial, my_sinus(t_initial, 4))
axs[1].stem(t_sample, my_sinus(t_sample, 4))
axs[1].set_xlabel("timp")
axs[1].set_ylabel("Amplitudine")



axs[2].plot(t_initial, my_sinus(t_initial, 16))
axs[2].stem(t_sample, my_sinus(t_sample, 16))
axs[2].set_xlabel("timp")
axs[2].set_ylabel("Amplitudine")

plt.show()
```

```python
# Exercitiul 3


def my_sinus(t, f=1):
    return np.sin(2 * np.pi * t * f)


t_initial = np.linspace(0, 1, 1000)
t_sample = np.linspace(0, 1, 34)

fig, axs = plt.subplots(3)
fig.subplots_adjust(hspace=1)

axs[0].plot(t_initial, my_sinus(t_initial))
axs[0].stem(t_sample, my_sinus(t_sample))
axs[0].set_title("f=1")
axs[0].set_xlabel("timp")
axs[0].set_ylabel("Amplitudine")


axs[1].plot(t_initial, my_sinus(t_initial, 4))
axs[1].stem(t_sample, my_sinus(t_sample, 4))
axs[1].set_title("f=4")
axs[1].set_xlabel("timp")
axs[1].set_ylabel("Amplitudine")


axs[2].plot(t_initial, my_sinus(t_initial, 16))
axs[2].stem(t_sample, my_sinus(t_sample, 16))
axs[2].set_xlabel("timp")
axs[2].set_ylabel("Amplitudine")

axs[2].set_title("f=16")

plt.show()
```
