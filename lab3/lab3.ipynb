```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Exercitiul 1

N = 8
F = np.zeros((N, N), dtype=np.complex_)
for i in range(N):
    for j in range(N):
        F[i][j] = np.power(np.e, -2j * np.pi * i * j / N)


fig, axs = plt.subplots(N, figsize=(12, 6), sharex=True, sharey=True)
fig.suptitle("DFT")

# Verificare matrice unitara
print(np.allclose(N * np.eye(N), np.dot(F, np.conj(F).T)))
# F * F^Transpusa = I(N)


for i in range(N):
    axs[i].plot([j + 1 for j in range(8)], F[i].real)
    axs[i].plot([j + 1 for j in range(8)], F[i].imag, ".--")
    axs[i].set_ylabel(f"i = {i}")
```

```python
# Exercitiul 2


t = np.linspace(0, 1, 1000)
omega = 6
semnal_sinus = np.sin(2 * np.pi * omega * t)
cerc_sinus = semnal_sinus * math.e ** -(2 * np.pi * 7 * 1j * t)

fig, axs = plt.subplots(1, 2, sharey=True)

fig.subplots_adjust(hspace=0.5)
axs[0].set_xlabel('Timp')
axs[0].set_ylabel('Amplitudine')
axs[0].plot(t, semnal_sinus)

axs[1].set_xlabel('Real')
axs[1].set_ylabel('Imaginar')
axs[1].set_aspect('equal')
axs[1].plot(np.real(cerc_sinus), np.imag(cerc_sinus))

plt.show()
```

```python
t = np.linspace(0, 1, 10000)
semnal_sinus = np.sin(2 * np.pi * 3 * t)
omegas = [3, 6, 8.5, 0.2]


def cerc_sinus(omega):
    return semnal_sinus * math.e ** -(2j * np.pi * t * omega)


fig, axs = plt.subplots(2, 2, sharey=True)

fig.subplots_adjust(hspace=0.5)

for i in range(2):
    for j in range(2):
        omega = omegas[i * 2 + j]
        complex_signal = cerc_sinus(omega)
        axs[i][j].scatter(np.real(complex_signal), np.imag(complex_signal), c=np.abs(semnal_sinus), cmap='viridis', s=0.1)
        axs[i][j].set_xlabel('Real')
        axs[i][j].set_ylabel('Imaginar')
        axs[i][j].set_title(f'ω = {omega}') 
        axs[i][j].set_xlim([-1.1, 1.1])
        axs[i][j].set_ylim([-1.1, 1.1])
        axs[i][j].axhline(0, color="black", linewidth=0.8)
        axs[i][j].axvline(0, color="black", linewidth=0.8)
        axs[i][j].scatter(np.real(np.mean(complex_signal)), np.imag(np.mean(complex_signal)), marker="o", color="black", s=50)



plt.show()
```
