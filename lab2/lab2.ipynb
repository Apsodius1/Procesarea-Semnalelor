```python
import numpy as np
import matplotlib.pyplot as plt 
import sounddevice as sd
import scipy
```

```python
# Exercitiul 1

# * Start Plot
fig, axs = plt.subplots(2, figsize=(10, 10))
plt.subplots_adjust(hspace=0.3)
fig.suptitle("Exercitiul 1")

# Functia sin
def sin(f0, t):
    return np.sin(2 * np.pi * f0 * t)

def cos(f0, t):
    return np.cos(2 * np.pi * f0 * t - np.pi/2)

nt1 = np.linspace(0, 1, 100000)

axs[0].plot(nt1, sin(1, nt1))
axs[0].set_title('Sin')
axs[0].set_xlabel("timp")
axs[0].set_ylabel("Amplitudine")
axs[0].set_xlim(0, 1)

axs[1].plot(nt1, cos(1, nt1))
axs[1].set_title('cos')
axs[1].set_xlabel("timp")
axs[1].set_ylabel("Amplitudine")
axs[1].set_xlim(0, 1)


# * Save plot
plt.savefig("exercitiul1.pdf")
plt.show()
```

```python
# Exercitiul 4
def sinFunction(f0, t):
    return np.sin(2 * np.pi * f0 * t)


def sawtooth(f0, t):
    return (f0 * t - np.floor(f0 * t + 1/2))


t = np.linspace(0, 1, 10000)
sinWave = sinFunction(1, t)
sawWave = sawtooth(1, t)
added = sinWave + sawWave

fig, axs = plt.subplots(3)
fig.suptitle("Semnale")

axs[0].plot(t, sinWave)
axs[1].plot(t, sawWave)
axs[2].plot(t, added)


for ax in axs:
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Amplitudine")

plt.show()
```

```python
# Exercitiul 6
def sinFunction(f0):
    t = np.linspace(0, 2, 10000)
    return np.sin(2 * np.pi * f0 * t)

fig, axs = plt.subplots(3)
fig.suptitle("Semnale")

axs[0].plot(t, sinFunction(1/2))
axs[1].plot(t, sinFunction(1/4))
axs[2].plot(t, sinFunction(0))


for ax in axs:
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Amplitudine")

plt.show()

```
